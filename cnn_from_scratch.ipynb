- How does your code resize the images (by cropping, stretching, etc)?  What size did you pick for the input tensor, and why?

<img src="static_images/icons/noun-answer-3361020.png" alt=">" style="width:25px"/> **Answer**: 
1. Resize and Crop:

The code uses transforms.Resize(256) to resize the images to a larger size (256x256).
It then applies transforms.RandomCrop(224, padding_mode="reflect", pad_if_needed=True) during training. This randomly crops a 224x224 patch from the resized image. The reflect padding mode reflects pixels at the edges to avoid creating black borders during cropping.
2. Data Augmentation (Training Only):

The training set uses several data augmentation techniques:
transforms.RandomAffine(scale=(0.9, 1.1), translate=(0.1, 0.1), degrees=10): This applies random affine transformations like scaling, translation, and rotation to the images, increasing the variability of the training data.
transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1): This randomly jitters the brightness, contrast, saturation, and hue of the images, further augmenting the data.
transforms.RandomHorizontalFlip: This randomly flips the images horizontally with a 50% chance.
3. Normalization:

Both training, validation, and test sets use transforms.ToTensor() to convert the PIL images to PyTorch tensors (CHW format).
Finally, transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]) is applied for normalization. These values are the commonly used ImageNet mean and standard deviation, suitable for pre-trained models trained on ImageNet.
4. Input Tensor Size (224x224):

The code resizes and crops the images to a final size of 224x224. This is a common input size for many pre-trained CNN architectures like ResNet and VGG.
5. Data Augmentation for Training:

Data augmentation is only applied to the training dataset. This helps to artificially increase the size and variability of the training data, leading to better model generalization and preventing overfitting.
6. No Augmentation for Validation and Test:

The validation and test sets do not use data augmentation. This is because we want to evaluate the model's performance on unseen data that hasn't been artificially modified.

Reasoning

The provided CNN architecture is a solid foundation for image classification tasks. It incorporates common building blocks like convolutional layers, pooling, activation functions, and fully connected layers. The use of batch normalization and dropout helps to improve training stability and generalization.

_Convolutional Layers_

Multiple convolutional layers with increasing filters: The architecture starts with convolutional layers with a small number of filters (16) and gradually increases the number of filters (32, 64, 128) in subsequent layers. This allows the network to learn increasingly complex features as it goes deeper.

Reasoning: This is a common practice in CNNs to capture hierarchical features. Lower layers learn simple features like edges and textures, while deeper layers learn more abstract features like shapes and objects.

Batch normalization after convolutional layers: Batch normalization is used to stabilize training and improve generalization by normalizing the activations.
Reasoning: This helps to reduce internal covariate shift, making the training process more stable and efficient.

Max pooling after convolutional layers: Max pooling is used to downsample the feature maps, reducing computational cost and creating a more robust representation.
Reasoning: This helps to reduce the dimensionality of the data while preserving important information.

ReLU activation function: ReLU is used as the activation function after each convolutional layer.
Reasoning: ReLU is a computationally efficient activation function that introduces non-linearity to the network, allowing it to learn complex patterns.

````python
batch_size = 32       # size of the minibatch for stochastic gradient descent (or Adam)
valid_size = 0.2       # fraction of the training data to reserve for validation
num_epochs = 50        # number of epochs for training
num_classes = 50       # number of classes. Do not change this
dropout = 0.3          # dropout for our model
learning_rate = 0.01  # Learning rate for SGD (or Adam)
opt = 'sgd'            # optimizer. 'sgd' or 'adam'
weight_decay = 1e-2 # regularization. Increase this to combat overfitting
momentum = 0.8
````

````python
from src.data import get_data_loaders
from src.train import optimize
from src.optimization import get_optimizer, get_loss
from src.model import MyModel

# get the data loaders using batch_size and valid_size defined in the previous
# cell
# HINT: do NOT copy/paste the values. Use the variables instead
data_loaders = get_data_loaders(batch_size=batch_size, valid_size=valid_size)# YOUR CODE HERE

# instance model MyModel with num_classes and drouput defined in the previous
# cell
model = MyModel(num_classes=num_classes,dropout=dropout) # YOUR CODE HERE

# Get the optimizer using get_optimizer and the model you just created, the learning rate,
# the optimizer and the weight decay specified in the previous cell
optimizer = get_optimizer(model=model,learning_rate=learning_rate,optimizer=opt,
                          weight_decay=weight_decay,momentum=momentum)
# YOUR CODE HERE

# Get the loss using get_loss
loss = get_loss()# YOUR CODE HERE

optimize(
    data_loaders,
    model,
    optimizer,
    loss,
    n_epochs=num_epochs,
    save_path="checkpoints/best_val_loss.pt",
    interactive_tracking=True
)
````
